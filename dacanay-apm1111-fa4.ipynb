```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution
a = 0  # minimum value
b = 1  # maximum value

# Generate x values
x = np.linspace(a - 0.1, b + 0.1, 100)

# Calculate the PDF values
pdf = np.where((x >= a) & (x <= b), 1/(b-a), 0)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, label='Uniform Distribution', color='blue')
plt.fill_between(x, pdf, color='lightblue', alpha=0.5)
plt.title('Continuous Uniform Distribution U(0, 1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.xlim(-0.1, 1.1)
plt.ylim(0, 1.2)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend()
plt.grid()
plt.show()
```
